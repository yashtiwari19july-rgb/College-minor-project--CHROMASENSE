```python
rows = int(input("Enter the number of rows for the pyramid: "))

for i in range(rows):
    # Print leading spaces
    for j in range(rows - i - 1):
        print(" ", end="")
    # Print stars
    for k in range(2 * i + 1):
        print("*", end="")
    # Move to the next line
    print()
```

```python
rows = int(input("Enter the number of rows for the left-weighted pyramid: "))

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
rows = int(input("Enter the number of rows for the right-weighted pyramid: "))

for i in range(1, rows + 1):
    # Print leading spaces
    for j in range(rows - i):
        print(" ", end="")
    # Print stars
    for k in range(i):
        print("*", end="")
    print()
```

### How the Right-Weighted Pyramid Code Works:

1.  **Get Input**: It first asks the user for the `rows` (height) of the pyramid.
2.  **Outer Loop (`for i in range(1, rows + 1)`)**: This loop iterates from `1` up to the specified `rows`. `i` represents the current row number.
3.  **Inner Loop for Spaces (`for j in range(rows - i)`)**: For each row `i`, this loop prints `rows - i` leading spaces. As `i` increases, the number of spaces decreases, pushing the stars to the right.
4.  **Inner Loop for Stars (`for k in range(i)`)**: After the spaces, this loop prints `i` stars. The number of stars increases with each row.
5.  **Newline (`print()`)**: After printing all spaces and stars for a row, an empty `print()` statement is used to move the cursor to the next line for the subsequent row.
